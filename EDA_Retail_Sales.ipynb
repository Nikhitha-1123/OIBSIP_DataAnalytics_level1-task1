```python
monthly_sales = df1['Total Amount'].resample('M').sum()
```

```python
monthly_sales.head()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(monthly_sales, marker='o', linestyle='-', color='blue')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales Amount")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
monthly_sales_rolling = monthly_sales.rolling(window=3).mean()

plt.figure(figsize=(10,5))
plt.plot(monthly_sales, label='Monthly Sales', alpha=0.6)
plt.plot(monthly_sales_rolling, label='3-Month Rolling Average', color='red', linewidth=2)
plt.title("Monthly Sales with Rolling Average")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.legend()
plt.grid(True)
plt.show()
```

### Time Series Insights:
- Sales peaked in [Month/Year] and dropped in [Month/Year].
- There is a clear seasonal trend indicating higher sales in [e.g., Q4].
- The rolling average shows a [rising/falling/stable] sales trend over time.
- Recommend preparing inventory and marketing campaigns accordingly.
